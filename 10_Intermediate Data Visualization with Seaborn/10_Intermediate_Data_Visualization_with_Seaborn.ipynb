## **Regression Plot In Seaborn**

| Feature        | lmplot                                      | regplot                                      |
|----------------|---------------------------------------------|----------------------------------------------|
| Function       | FacetGrid with regression plots              | Scatter plot with regression line             |
| Usage          | Explore relationships across multiple plots | Examine relationship between two variables    |
| Data           | Can handle wide-form and long-form datasets | Requires a single DataFrame                  |
| Plot Type      | Supports various plot types (e.g., scatter, line, etc.) | Mainly used for scatter plots with regression line |
| Multiple Plots | Automatically creates subplots               | Requires explicit subplot creation            |
| Categorical    | Handles categorical variables (hue, col, row) | Limited support for categorical variables     |
| Flexibility    | Offers extensive customization options       | Limited customization options                |
| Regression Fit | Provides confidence intervals and bootstrapped estimates | Provides confidence intervals only            |
| Integration    | Integrated with other Seaborn functions      | Standalone function                          |

```python
sns.regplot(x='insurance_losses', y='premiums', data=insurance)
plt.show()
```

```python
sns.lmplot(x='insurance_losses', y='premiums', data=insurance, height=5, aspect=2)
plt.show()
```

# 2Ô∏è‚É£**Customizing Seaborn Plots**


## **Using Seaborn Styles**


#### **Setting the default style**


```python
# Plot the pandas histogram
plt.figure(figsize=(10, 5))
fy18['fmr_2'].plot.hist()
plt.show()
```

```python
# Set the default seaborn style: darkgrid
sns.set()
plt.figure(figsize=(10, 5))

# Plot the pandas histogram again
fy18['fmr_2'].plot.hist()
```

#### **Comparing styles**


```python
sns.set_style('dark')
plt.figure(figsize=(10, 5))
sns.histplot(fy18['fmr_2'], kde=True)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 5))
sns.histplot(fy18['fmr_2'], kde=True)
plt.show()
```

#### **Removing spines**

In general, visualizations should minimize extraneous markings so that the data speaks for itself. Seaborn allows you to remove the lines on the top, bottom, left and right axis, which are often called spines.


```python
# Ste the style to white
sns.set_style('white')
plt.figure(figsize=(10, 5))

# Create a regression plot
sns.lmplot(data=fy18,
          x='pop2010',
          y='fmr_2',
          height=5,
          aspect=2)

# Remove the spines
# Spines are the lines that border the plot area. By default, they are present on the top and right sides of the plot.
sns.despine(right=True, top=True)
plt.show()
```

## **Colors in Seaborn**
#### **Matplotlib color codes**


```python
# Set styles, enable color code, and create a magenta distplot
sns.set(color_codes=True)
plt.figure(figsize=(10, 5))
sns.displot(fy18['fmr_3'], kde=True, color='m')
plt.show()
```

#### **Using default palettes**


```python
# Loop through differences between bright and colorblind palettes
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    plt.figure(figsize=(10, 5))
    sns.displot(fy18['fmr_3'], kde=True)
    plt.show()
```

#### **Creating Custom Palettes**



```python
sns.palplot(sns.color_palette('Purples', 8))
plt.show()
```

```python
sns.palplot(sns.color_palette('husl', 10))
plt.show()
```

```python
sns.palplot(sns.color_palette('coolwarm', 6))
plt.show()
```

## **Customizing with matplotlib**
#### **Using matplotlib axes**


```python
sns.set_palette('tab20')

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the distribution of data
sns.histplot(fy18['fmr_3'], ax=ax, kde=True)

# Create a more descriptive x axis label
ax.set(xlabel="3 Bedroom Fair Market Rent",
       ylabel="Number of count")

plt.show()
```

#### **Additional plot customizations**


```python
# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the distribution of 1 bedroom rents
sns.histplot(fy18['fmr_1'], ax=ax, kde=True)

# Modify the properties of the plot
ax.set(xlabel='1 Bedroom Fair Market Rent',
       xlim=(100, 1500),
       title='US Rent')

plt.show()
```

#### **Adding annotations**


```python
# Create a figure and axes, Then plot the data
fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(fy18['fmr_1'], ax=ax, kde=True)

# Customize the labels and limits
ax.set(xlabel='1 Bedroom Fair Market Rent', 
       xlim=(100, 1500),
       title='US Rent')

# Add vertical lines for the median and mean
ax.axvline(x=fy18['fmr_1'].median(),
           color='r',
           label='Median',
           linestyle='--',
           linewidth=2)

ax.axvline(x=fy18['fmr_1'].mean(),
           color='g',
           label='Mean',
           linestyle='-',
           linewidth=2)

# Show the legend
ax.legend()

plt.show()
```

#### **Multiple plots**

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(15, 5))

# Plot the distribution of 1 bedroom apartments on ax0
sns.histplot(fy18['fmr_1'], ax=ax0, kde=True)
ax0.set(xlabel='1 Bedroom Fair Market Rent', xlim=(100, 1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.histplot(fy18['fmr_2'], ax=ax1, kde=True)
ax1.set(xlabel='2 Bedroom Fair Market Rent', xlim=(100, 1500))

plt.show()
```

## **Matrix plots**

#### **Creating heatmaps**

```python
# Create a crosstab table of the data
pd_crosstab = pd.crosstab(daily_show['Group'], daily_show['YEAR'])
display(pd_crosstab)

# Plot a heatmap of the table
sns.heatmap(pd_crosstab, linewidths=0.1)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```

# 4Ô∏è‚É£**Creating Plots on Data Aware Grids**

## **Using `FacetGrid`, `catplot` and `lmplot`**



#### **Building a FacetGrid**


```python
# Create FacetGrid with Degree_Type and specify the order of the rows usiung row_order
g2 = sns.FacetGrid(college_data, row='Degree_Type',
                  row_order=['Graduate', 'Bachelors', 'Associates', 'Certificate'])

# Map a pointplot of SAT_AVG_ALL onto the grid
g2.map(sns.pointplot, 'SAT_AVG_ALL', order=['Graduate', 'Bachelors', 'Associates', 'Certificate'])

plt.show()
```

#### **Using a catplot**


```python
# Create a factorplot(catplot in latest seaborn) that contains boxplots of Tuition values
sns.catplot(data=college_data,
              x='Tuition',
              kind='box',
              row='Degree_Type')

plt.show()
```

```python
# Create a facetted pointplot of Average SAT_AVG_ALLl scores facetted by Degree_Type
sns.catplot(data=college_data,
              x='SAT_AVG_ALL',
              kind='point',
              row='Degree_Type',
              row_order=['Graduate', 'Bachelors', 'Associates', 'Certificate'])

plt.show()
```

#### **Using a lmplot**

```python
# Create a FacetGrid varying by column and columns ordered with the degree_order variable
degree_ord = ['Graduate', 'Bachelors', 'Associates']
g = sns.FacetGrid(college_data, col='Degree_Type', col_order=degree_ord)

# Map a scatterplot of Undergrad Population compared to PCTPELL
g.map(plt.scatter, 'UG', 'PCTPELL')

plt.show()
```

```python
# Re-create the plot above as an lmplot
sns.lmplot(data=college_data,
          x='UG',
          y='PCTPELL',
          col='Degree_Type',
          col_order=degree_ord)

plt.show()
```

```python
# Create an lmplot that has a column for Ownership, a row for Degree_Type and hue based on the WOMENONLY column
inst_ord = ['Public', 'Private non-profit']

sns.lmplot(data=college_data,
          x='SAT_AVG_ALL',
          y='Tuition',
          col='Ownership',
          row='Degree_Type',
          row_order=['Graduate', 'Bachelors'],
          hue='WOMENONLY',
          col_order=inst_ord)

plt.show()
```

## **Using PairGrid and pairplot**



#### **Building a PairGrid**


```python
# Create a PairGrid with a scatter plot for fatal_collisions and premiums
g = sns.PairGrid(insurance, vars=["fatal_collisions", "premiums"])
g2 = g.map(sns.scatterplot)

plt.show()
```

```python
# Create the same PairGrid but map a histogram on the diag
g = sns.PairGrid(insurance, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(sns.histplot)
g3 = g2.map_offdiag(sns.scatterplot)

plt.show()
```

#### **Using a pairplot**


```python
# Create a pairwise plot of the variables using a scatter plot
sns.pairplot(insurance,
            vars=['fatal_collisions', 'premiums'],
             kind='scatter'
            )

plt.show()
```

```python
# Plot the same data but use a different color palette and color code by Region
sns.pairplot(insurance,
            vars=['fatal_collisions', 'premiums'],
            kind='scatter',
            hue='Region',
            palette='RdBu',
            diag_kws={'alpha':.5})

plt.show()
```

#### **Additional pairplots**

```python
# Build a pairplot with different x and y variables
sns.pairplot(data=insurance,
            x_vars=['fatal_collisions_speeding', 'fatal_collisions_alc'],
            y_vars=['premiums', 'insurance_losses'],
            kind='scatter',
            hue='Region',
            palette='husl')

plt.show()
```

```python
# plot relationships between insurance_losses and premiums
sns.pairplot(data=insurance,
            vars=['insurance_losses', 'premiums'],
            kind='reg',
            palette='BrBG',
            diag_kind='kde',
            hue='Region')

plt.show()
```

## **Using JointGrid and jointplot**


#### **Building a JointGrid and jointplot**


```python
# Build a JointGrid comparing humidity and total_rentals
sns.set_style("whitegrid")
g = sns.JointGrid(x="hum",
                  y="total_rentals",
                  data=bike_share,
                  xlim=(0.1, 1.0))

g.plot(sns.regplot, sns.histplot)

plt.show()
```

```python
# Create a jointplot similar to the JointGrid
sns.jointplot(x="hum",
             y="total_rentals",
             kind="reg",
             data=bike_share)

plt.show()
```

#### **Jointplots and regression**


```python
# Plot temp vs. total_rentals as a regression plot
sns.jointplot(x='temp',
             y='total_rentals',
             kind='reg',
             data=bike_share,
             order=2,
             xlim=(0, 1))
```

```python
# Plot a jointplot showing the residuals
sns.jointplot(x='temp',
             y='total_rentals',
             kind='resid',
             data=bike_share,
             order=2)
```

#### **Complex jointplots**

```python
# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = sns.jointplot(x='temp', y='casual',
                  kind='scatter',
                  data=bike_share,
                  marginal_kws=dict(bins=10, rug=True))

# Overlay a kdeplot on the scatter plot
g.plot_joint(sns.kdeplot)

plt.show()
```

```python
# Replicate the above plot but only for registered riders
# Create a jointplot of temp vs. registered riders
g = sns.jointplot(x='temp', y='registered',
                  kind='scatter',
                  data=bike_share,
                  marginal_kws=dict(bins=10, rug=True))

# Overlay a kdeplot on the scatter plot
g.plot_joint(sns.kdeplot)

plt.show()
```
