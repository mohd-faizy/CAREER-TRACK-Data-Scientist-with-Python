### **Using outer join to select actors**
One cool aspect of using an `outer` join is that, because it returns all rows from both `merged` tables and `null` where they `do not match`, you can use it to find rows that do not have a match in the other table.

To try for yourself, you have been given two tables with a list of actors from two popular movies: `Iron Man 1` and `Iron Man 2`. Most of the actors played in both movies.

>**Use an outer join to find actors who did not act in both movies.**

![outer join](https://assets.datacamp.com/production/repositories/5486/datasets/c5d02ebba511e90ae132f89ff091e6729c040bd2/noJoin.png)

````python
movies = pd.read_pickle("/content/movies.p")
movies = movies.rename(columns={'id':'movie_id'})

cast = pd.read_pickle("/content/casts.p")
movies_cast = cast.merge(movies, on='movie_id', how='outer')

# movies_cast[movies_cast['title'].isin(['Iron Man', 'Iron Man 2', 'Iron Man 3'])]
iron_1_actors = movies_cast[movies_cast['title'].isin(['Iron Man'])]
iron_2_actors = movies_cast[movies_cast['title'].isin(['Iron Man 2'])]

# Merge iron_1_actors to iron_2_actors on id with outer join using suffixes
iron_1_and_2 = iron_1_actors.merge(iron_2_actors,
                                   on='id',
                                   how='outer',
                                   suffixes=('_1', '_2'))

# Create an index that returns true if name_1 or name_2 are null
m = ((iron_1_and_2['name_1'].isnull()) | (iron_1_and_2['name_2'].isnull()))

# Print the first few rows of iron_1_and_2
print(iron_1_and_2[m].head())
````

### **Self join**

Merging a table to itself can be useful when you want to compare values in a column to other values in the same column.

In this exercise, you will practice this by creating a table that for each `movie` will list the `movie director` and a member of the `crew` on one row. You have been given a table called `crews`, which has columns `id`, `job`, and `name`.

First, merge the table to itself using the `movie ID`. This merge will give you a larger table where for each movie, every job is matched against each other. Then select only those rows with a director in the left table, and avoid having a row where the director's job is listed in both the left and right tables. This filtering will remove job combinations that aren't with the director.

````python
crews = pd.read_pickle("/content/crews.p")
````

````python
# Merge the crews table to itself
crews_self_merged = crews.merge(crews, on='id', how='inner', suffixes=('_dir', '_crew'))

# Create a boolean index to select the appropriate rows
boolean_filter = ((crews_self_merged['job_dir'] == 'Director') & (crews_self_merged['job_crew'] != 'Director'))
direct_crews = crews_self_merged[boolean_filter]

# Print the first few rows of direct_crews
display(direct_crews.head())
````

By merging the table to itself, you compared the value of the `__director__` from the jobs column to other values from the jobs column. With the output, you can quickly see different movie directors and the people they worked with in the same movie.

### **Semi-join**
Steps to perform a semi-join:

It returns only the rows from the left table that have a match in the right table.

- Merge the `left` and `right` tables on `key` column using an `inner` join. 
- Search if `key` column in the `left` table is in the merged tables using `.isin()` method creating Boolean Series 
- Subset the rows of the `left` table. 




````python
# Create two dataframes
df1 = pd.DataFrame({'key': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'key': [2, 3, 4], 'age': [25, 30, 35]})

# Merge the two dataframes on the 'id' column
merged_df = df1.merge(df2[['key']], on='key', how='inner', indicator=True)
display(merged_df)

# Get the names of the people with an age
with_age_df = merged_df[['key', 'name']]
print("The following people have an age:\n", with_age_df)
````

````python
# Create example DataFrames
df1 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'category': ['A', 'B', 'B', 'C', 'C']})
df2 = pd.DataFrame({'id': [2, 4], 'value': [10, 20]})

# Use isin() to subset df1 to rows with id in df2
df1_subset = df1[df1['id'].isin(df2['id'])]

# Group df1_subset by category and count the id rows
count_by_category = df1_subset.groupby('category', as_index=False).agg({'id': 'count'})

# Merge the resulting DataFrame with a third DataFrame on the category column
df3 = pd.DataFrame({'category': ['A', 'B', 'C'],
                    'name': ['apple', 'banana', 'carrot']})

result = count_by_category.merge(df3, on='category')

print(result)
````

### **Anti join**
This type of join that returns all the rows from the left table that don't have a match in the right table.



````python
# Create two dataframes
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'id': [2, 3, 4], 'age': [25, 30, 35]})

# Merge the two dataframes on the 'id' column
merged_df = df1.merge(df2, on='id', how='left', indicator=True)
display(merged_df)

# Select the rows where _merge is left_only
left_only_df = merged_df.loc[merged_df['_merge'] == 'left_only', :]

# Get the names of the people who don't have an age
display(left_only_df['name'])
````

ou performed an anti join by first merging the tables with a left join, selecting the ID of those employees who did not support a top customer, and then subsetting the original employee's table. From that, we can see that there are five employees not supporting top customers. Anti joins are a powerful tool to filter a main table

### **Concatenation basics**

````python
# Dummy data with non-numeric values
data1 = pd.DataFrame({'Name': ['John', 'Alice', 'Bob'],
                      'Age': [30, 25, 27],
                      'Gender': ['Male', 'Female', 'Male'],
                      'Occupation': ['Doctor', 'Engineer', 'Teacher']})

data2 = pd.DataFrame({'Name': ['Mary', 'Jane', 'Liam'],
                      'Age': [35, 29, 31],
                      'Gender': ['Female', 'Female', 'Male'],
                      'Occupation': ['Doctor', 'Nurse', 'Engineer']})

data3 = pd.DataFrame({'Name': ['Sophie', 'David', 'Eric'],
                      'Age': [24, 32, 28],
                      'Gender': ['Female', 'Male', 'Male'],
                      'Occupation': ['Teacher', 'Doctor', 'Lawyer']})

# Concatenate the dataframes
concatenated_data = pd.concat([data1, data2, data3], sort=True)
print("concatenated_data:\n", concatenated_data)

# Concatenate the dataframes with reset index
concatenated_data_reset_index = pd.concat([data1, data2, data3], ignore_index=True, sort=True)
print("\nconcatenated_data_reset_index:\n", concatenated_data_reset_index)

# Concatenate the dataframes and keep only common columns
concatenated_data_common_columns = pd.concat([data1, data2, data3], join='inner', sort=True)
print("\nconcatenated_data_common_columns:\n", concatenated_data_common_columns)
````

### **`merge_asof()` and `merge_ordered()` differences**

| Feature| `merge_asof()` | `merge_ordered()` |
| ------------- | ----------------------------------- | ----------------- |
| Matching      | Matches on closest key              | Matches on exact key             |
| Direction     | Forward or backward                 | Forward, backward, or nearest         |
| Sorting       | Must be sorted on the key column(s) | Must be sorted on the key column(s)   |
| Overlapping   | Overlapping values are included     | Overlapping values can be included/excluded depending on `fill_method` argument |
| Interpolation | Can interpolate missing values      | Can interpolate missing values using `method` argument |

### **Using `.melt()` to reshape government data**


`pandas.melt()` is used to reshape your data  from wide to long **for easier analysis or visualization**. 


```python
pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None, ignore_index=Tr
```


Pandas`.melt()` is a function that can be used to transform a DataFrame from **wide format to long format**.

It is useful for converting a DataFrame into a format where one or more columns are identifier variables (`id_vars`), while all other columns, considered measured variables (`value_vars`), are `unpivoted` to the row axis, leaving just two non-identifier columns, `variable` and `value`.

````python
# Create a sample DataFrame
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
display(df)

# Melt the DataFrame
melted_df = pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])

# Print the melted DataFrame
display(melted_df)
````

In this example, we specify `column A` as the `identifier variable` (`id_vars`) and `columns B` and `C` as the `measured variables` (`value_vars`). The resulting melted DataFrame has three columns: `A`, `variable`, and `value`. The variable column contains the column names from the original DataFrame (`B` and `C`) and the value column contains the corresponding values.

````python
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})

# Melt the DataFrame
melted_df = pd.melt(df, id_vars=['A'], value_vars=['B', 'C'], var_name='Variable', value_name='Value')

# Print the melted DataFrame
display(melted_df)
````
