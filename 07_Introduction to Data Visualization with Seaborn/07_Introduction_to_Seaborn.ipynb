### **Customizing point plots**


**`Point plots` vs. `line plots`**

**Both show:**
- Mean of quantitative variable
- 95% confidence intervals for the mean

**Differences:**
- Line plot has quantitative variable (usually time) on x-axis
- Point plot has categorical variable on x-axis

**`Point plots` vs. `bar plots`**

**Both show:**
- Mean of quantitative variable
- 95% confidence intervals for the mean

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point")

# Show plot
plt.show()
```

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2)

# Show plot
plt.show()
```

By setting `capsize=0.2`, you are specifying that the length of the caps should be 0.2 times the width of the point estimate. This parameter allows you to control the visual appearance of the caps in relation to the point estimate.

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
            data=student_data,
            kind="point",
            capsize=0.2,
            join=False)

# Show plot
plt.show()
```

hile the average number of absences is slightly smaller among students with higher-quality family relationships, the large confidence intervals tell us that we can't be sure there is an actual association here.

### **`FacetGrids` vs. `AxesSubplots`**

#### **FacetGrids:**

- FacetGrids are part of the seaborn library and are specifically designed for creating grids of plots based on one or more categorical variables.
- FacetGrids allow you to easily create a grid of subplots, where each subplot  represents a unique combination of the categories in the specified variables.
- FacetGrids are especially useful when you want to visualize relationships - between variables across different subsets of your data.
- FacetGrids provide built-in capabilities for handling categorical variables,  including options for row and column facets, and hue variables for color - encoding.
- They provide additional functionality such as the ability to apply different  plot types to each subplot, apply statistical functions, and more.


#### **AxesSubplots:**

- AxesSubplots are part of the matplotlib library and provide a way to manually create and arrange subplots in a single figure.
- AxesSubplots give you more fine-grained control over the layout and - customization of individual subplots within a figure.
- They are suitable for creating a variety of plot layouts, including grids, - rows, columns, or any custom arrangement you desire.
- With AxesSubplots, you need to explicitly define the positions and sizes of - each subplot in the figure, which gives you more flexibility but requires more  manual adjustments.

|Object|Plot type|Characteristics|
|------|---------|---------------|
|`FacetGrid`|`relplot()`, `catplot()`|Can create Subplots|
|`AxesSubplot`|`scatterplot()`, `countplot()` etc|Only creates a single plot|

```python
plt.style.use('ggplot')

# Load the tips dataset
tips_data = sns.load_dataset("tips")

# Create a FacetGrid with two facets based on the "time" variable
g = sns.FacetGrid(data=tips_data, col="time")

# Map a scatter plot onto the FacetGrid
g.map(sns.scatterplot, "total_bill", "tip")

# Show the plot
plt.show()
```

```python
# AxesSubplots
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2)

# Generate random data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot data on the first subplot
axs[0].plot(x, y1)
axs[0].set_title('Sin(x)')

# Plot data on the second subplot
axs[1].plot(x, y2)
axs[1].set_title('Cos(x)')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
